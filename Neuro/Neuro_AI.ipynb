```python
import getpass
import openai
import os
#   API        
openai_key = getpass.getpass("OpenAI API Key:")
os.environ["OPENAI_API_KEY"] = openai_key
openai.api_key = openai_key
```

```python
#      API OpenAI
result = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
        {"role": "system", "content": "  .     Python"},
        {"role": "user", "content": "   Python?   ?"}
    ]
)

print(result)  #   
message = result['choices'][0]['message']['content']
print(message)
```

```python
result = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
        {"role": "system", "content": "-    ,        - "},
        {"role": "user", "content": "      ?"}
    ]
)

print(result)  #   
message = result['choices'][0]['message']['content']
print(message)
```

```python
result = openai.ChatCompletion.create(
  model="gpt-4o",
  messages=[
        {"role": "system", "content": "- ,   "},
        {"role": "user", "content": "      ?"}
    ]
)

print(result)  #   
message = result['choices'][0]['message']['content']
print(message)
```

```python
result = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
        {"role": "system", "content": "-  ,   ,     ,     "},
        {"role": "user", "content": "      ?"}
    ],
  max_tokens=100,
  temperature=0.5,
  n=1
  )

print(result)  #   
message = result['choices'][0]['message']['content']
print(message)
```

```python
result = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
        {"role": "system", "content": "-  ,   ,     ,     "},
        {"role": "user", "content": "      ?"}
    ],
  max_tokens=300,
  temperature=1,
  n=3
)

print(result)  #   
message_0 = result['choices'][0]['message']['content']
print("  :")
print(message_0)
print("  :")
message_1 = result['choices'][1]['message']['content']
print(message_1)
print("  :")
message_2 = result['choices'][2]['message']['content']
print(message_2)
```
